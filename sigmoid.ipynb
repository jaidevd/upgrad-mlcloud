```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
x = np.random.normal(0.5, 1, size=(40,))
x.sort()
pos = x[x > 0.5]
neg = x[x < 0.5]
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(pos, np.zeros(pos.shape), c='g', marker='x', linewidth=3, s=100)
ax.scatter(neg, np.zeros(neg.shape), c='r', marker='o', linewidth=3, s=100)

# ax.plot(x, 1 / (1 + np.exp(-x)))
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(pos, np.zeros(pos.shape), c='g', marker='x', linewidth=3, s=100, label='+ve class')
ax.scatter(neg, np.zeros(neg.shape), c='r', marker='o', linewidth=3, s=100, label='-ve class')
ax.plot(neg, np.zeros(neg.shape), 'b', alpha=0.3, label='Step Function')
ax.plot(pos, np.ones(pos.shape), 'b', alpha=0.3)
ax.vlines(0.5, 0, 1, 'b', alpha=0.3)
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(pos, np.zeros(pos.shape), c='g', marker='x', linewidth=3, s=100, label='+ve class')
ax.scatter(neg, np.zeros(neg.shape), c='r', marker='o', linewidth=3, s=100, label='-ve class')
# ax.plot(x + 0.5, 1 / (1 + np.exp(-x - 0.5)), 'b', alpha=0.3, label='Sigmoid / Logistic Function')
xx = np.linspace(-3, 3, 100)
yy = 1 / (1 + np.exp(-xx))
ax.plot(xx +0.5, yy, 'b', alpha=0.3, label='Sigmoid / Logistic Function')
ax.hlines(0.5, -2, 0.5, 'k', linestyle='dashed', linewidth=2)
ax.vlines(0.5, 0, 0.5, 'k', linestyle='dashed', linewidth=2)
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(pos, np.ones(pos.shape), c='g', marker='x', linewidth=3, s=100, label='+ve class')
ax.scatter(neg, np.zeros(neg.shape), c='r', marker='o', linewidth=3, s=100, label='-ve class')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
plt.legend()
```
